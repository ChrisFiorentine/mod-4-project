```python
columns = ['Date', 'Minutes']
minutes_df = pd.DataFrame(columns = columns)
```

```python
data = []
for date in meeting_dates:
    pdf_file = open('{}.pdf'.format(date.replace('/','-')), 'rb')
    pdf_reader = PyPDF2.PdfFileReader(pdf_file)
    text = ''
    for num in range(0,pdf_reader.numPages):
        page_obj = pdf_reader.getPage(num).extractText().replace('\n','')
        text = text + page_obj
    data.append([date,text])
    print(date)
minutes_df = pd.DataFrame(data, columns=columns)
    
    
```

```python
import datetime
minutes_df['Date'] = pd.to_datetime(minutes_df['Date'])
minutes_df['Release Date'] = minutes_df['Date'] + pd.DateOffset(21)
minutes_df['End Date'] = minutes_df['Release Date'] + pd.DateOffset(28)
minutes_df.head()
```

```python
stock_prices = pd.read_csv('file:///Users/chrisfiorentine/Downloads/HistoricalPrices%20(1).csv')
stock_prices
```

```python
stock_prices['Date'] = pd.to_datetime(stock_prices['Date'])
```

```python
stock_prices.rename(columns = {'Date':'Stock_Date'}, inplace = True)
```

```python
stock_prices.head()
```

```python
stock_prices.set_index('Stock_Date', inplace = True)
```

```python
def sandp500_return(row):
    start_price = stock_prices[stock_prices['Stock_Date'] == row['Release Date']][' Open']
    end_price = stock_prices[stock_prices['Stock_Date'] == row['End Date']][' Close']
    r = ((end_price-start_price)/start_price)
    return r
```

```python
minutes_df['Return'] = minutes_df.apply(lambda x: sandp500_return(x), axis = 1)
```
